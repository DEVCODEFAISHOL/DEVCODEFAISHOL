```python
import pandas as pd
import numpy as np
import random
from openpyxl import Workbook
from openpyxl.utils.dataframe import dataframe_to_rows
from openpyxl.styles import Alignment, Border, Side, Font


def generate_custom_scores_excel_report(filename="TUGAS_REYHAN_INFORMATIKA.xlsx"):
    """Generates a styled report with custom header, sorted names, custom scores for Reyhan, and saves it to an Excel file."""

    names = [
        "ABDUL MUIZ",
        "AHMAD YUDI",
        "ALDI PRATAMA",
        "ARIPIN ILHAM",
        "AZZAM",
        "BADAL FIRMAN",
        "DAFFA HANUNG LAKSANA",
        "DANAR WISNU AJI",
        "DIMAS TAUFIQ HIDAYAH",
        "EDWIN FIRMANSYAH",
        "EVAN BEDRI PRANAMA",
        "FADLAN PETER H.",
        "FAHRI",
        "FEBRIANSYAH",
        "FERDIANSYAH",
        "IRVANDA BAGAS S.",
        "M.ALI KADAFI",
        "M.ARYA IMANULLAH",
        "M.DEWA SAPOSETRA",
        "M.FADIL WAHYU.P",
        "M.REYHAN",
        "MUSLIM AL BAHKRI",
        "RAFA WAHAB IBRANI",
        "RAKA ARYANSAH",
        "RAKA ATHALLAH PRADIPTA",
        "RANDI MUHAMMAD",
        "RENDI SUBAHTIAR",
        "RETNO WULAN DARI",
        "RIA PERMATA SARI",
        "RIKO MAULANA",
        "RISKI HIDAYAT",
        "SITI FATIMATUL ZAHRA",
        "SULAIMAN BAROKAH",
        "SYARIF HIDAYATULLAH"
    ]

    sorted_names = sorted(names)
    num_students = len(sorted_names)

    # Generate random scores for all students first
    data = {
        "no": list(range(1, num_students + 1)),  # Sequential numbering
        "NAMA": sorted_names,
        "B.INGGRIS": [random.randint(70, 90) for _ in range(num_students)],
        "B.INDONESIA": [random.randint(70, 90) for _ in range(num_students)],
        "INFORMATIKA": [random.randint(70, 90) for _ in range(num_students)],
        "matematika": [random.randint(70, 90) for _ in range(num_students)],
        "IPAS": [random.randint(70, 90) for _ in range(num_students)]
    }

    df = pd.DataFrame(data)

    # Set custom scores for Muhammad Reyhan
    reyhan_index = df[df["NAMA"] == "M.REYHAN"].index[0]
    df.loc[reyhan_index, "B.INGGRIS"] = 82
    df.loc[reyhan_index, "B.INDONESIA"] = 87
    df.loc[reyhan_index, "INFORMATIKA"] = 82
    df.loc[reyhan_index, "matematika"] = 82
    df.loc[reyhan_index, "IPAS"] = 83

    df["JUMLAH"] = df[["B.INGGRIS", "B.INDONESIA", "INFORMATIKA", "matematika", "IPAS"]].sum(axis=1)
    df["RATA RATA"] = df["JUMLAH"] / 5
    df["RATA RATA"] = df["RATA RATA"].round(1)

    df_ranked = df.sort_values(by="RATA RATA", ascending=False).reset_index(drop=True)
    df_ranked["RANGKING"] = df_ranked.index + 1

    # Reorder columns to match the original excel structure
    df_ranked = df_ranked[["no", "NAMA", "B.INGGRIS", "B.INDONESIA", "INFORMATIKA", "matematika", "IPAS", "JUMLAH", "RATA RATA", "RANGKING"]]

    wb = Workbook()
    ws = wb.active

    # Custom Header Information
    header_lines = [
        ["", "", "", "", "", "", "", "", "", ""],
        ["", "", "", "", "", "", "", "", "", ""],
        ["Muhammad reyhan", "", "", "", "", "", "", "", "", ""],
        ["kelas x TMI a", "", "", "", "", "", "", "", "", ""],
        ["pelajaran informatika", "", "", "", "", "", "", "", "", ""],
        ["", "", "", "", "", "", "", "", "", ""],
        ["", "", "", "", "", "", "", "", "", ""],
        ["no", "NAMA", "materi", "", "", "", "JUMLAH", "RATA RATA", "RANGKING"],
        ["", "", "B.INGGRIS", "B.INDONESIA", "INFORMATIKA", "matematika", "IPAS", "", "", ""],
    ]

    for row_idx, row_data in enumerate(header_lines):
        for col_idx, cell_value in enumerate(row_data):
            ws.cell(row=row_idx + 1, column=col_idx + 1, value=cell_value)

    # Add data from DataFrame
    for row in dataframe_to_rows(df_ranked, index=False, header=False):
        ws.append(row)

    # Define border style
    thin_border = Border(left=Side(style='thin'),
                       right=Side(style='thin'),
                       top=Side(style='thin'),
                       bottom=Side(style='thin'))

    # Apply border to all cells
    for row in ws.iter_rows(min_row=8, max_row=ws.max_row):
        for cell in row:
            cell.border = thin_border

    # Align all cells to center
    for row in ws.iter_rows():
        for cell in row:
            cell.alignment = Alignment(horizontal='center', vertical='center')

    # Adjust column widths (adjust values as needed)
    ws.column_dimensions['A'].width = 5
    ws.column_dimensions['B'].width = 20
    ws.column_dimensions['C'].width = 12
    ws.column_dimensions['D'].width = 12
    ws.column_dimensions['E'].width = 12
    ws.column_dimensions['F'].width = 12
    ws.column_dimensions['G'].width = 12
    ws.column_dimensions['H'].width = 10
    ws.column_dimensions['I'].width = 10
    ws.column_dimensions['J'].width = 10

    # Set Headers to Bold
    header_row = ws[8]
    for cell in header_row:
        cell.font = Font(bold=True)
    header_row_2 = ws[9]
    for cell in header_row_2:
        cell.font = Font(bold=True)

    #Save the file
    wb.save(filename)

    print(f"Report generated and saved to '{filename}'")

# Generate the excel report
generate_custom_scores_excel_report()
```
